```python
import graphviz

dot = graphviz.Digraph("Dynastia Windsorów")
dot.attr(label='Dynastia Windsorów', labelloc='t', fontsize='20')

dot.node('George_V', 'George V\n1865-1936', style='filled', color='lightgreen')
dot.node('Edward_VIII', 'Edward VIII\n1894-1972', style='filled', color='lightgreen')
dot.node('George_VI', 'George VI\n1895-1952', style='filled', color='lightgreen')
dot.node('Elizabeth_II', 'Elizabeth II\n1926-2022', style='filled', color='lightgreen')
dot.node('Charles_III', 'Charles III\n1948-', style='filled', color='lightgreen')
dot.node('Anne', 'Anne\n1950-', style='filled', color='lightgreen')
dot.node('Andrew', 'Andrew\n1960-', style='filled', color='lightgreen')
dot.node('Edward', 'Edward\n1964-', style='filled', color='lightgreen')
dot.node('William', 'William\n1982-', style='filled', color='lightgreen')
dot.node('Harry', 'Harry\n1984-', style='filled', color='lightgreen')
dot.node('George', 'George\n2013-', style='filled', color='lightgreen')
dot.node('Charlotte', 'Charlotte\n2015-', style='filled', color='lightgreen')
dot.node('Louis', 'Louis\n2018-', style='filled', color='lightgreen')
dot.node('Archie', 'Archie\n2019-', style='filled', color='lightgreen')
dot.node('Lilibet', 'Lilibet\n2021-', style='filled', color='lightgreen')

dot.edge('George_V', 'Edward_VIII')
dot.edge('George_V', 'George_VI')
dot.edge('George_VI', 'Elizabeth_II')
dot.edge('Elizabeth_II', 'Charles_III')
dot.edge('Elizabeth_II', 'Anne')
dot.edge('Elizabeth_II', 'Andrew')
dot.edge('Elizabeth_II', 'Edward')
dot.edge('Charles_III', 'William')
dot.edge('Charles_III', 'Harry')
dot.edge('William', 'George')
dot.edge('William', 'Charlotte')
dot.edge('William', 'Louis')
dot.edge('Harry', 'Archie')
dot.edge('Harry', 'Lilibet')

dot.render('windsor_dynasty', format='png', cleanup=True)

from IPython.display import Image
Image('windsor_dynasty.png')
```
